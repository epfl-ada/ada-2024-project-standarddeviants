```python
heatmap_df = (
    df.value_counts(subset=["Institution", "Target Name"])
    .reset_index()
    .pivot_table(values="count", index="Institution", columns="Target Name")
    .fillna(0)
    .reindex(df["Institution"].value_counts().index.tolist())
)
heatmap_df = heatmap_df[
    df["Target Name"].value_counts().index.intersection(heatmap_df.columns)
]
heatmap_df
```

```python
sns.heatmap(heatmap_df.iloc[:20, :20])
```

```python
# https://blog.ddavo.me/posts/tutorials/seaborn-heatmap-marginal-distribution/
grid = sns.JointGrid()

top_n_institutions = 20
top_n_targets = 100

joint_heatmap_df = heatmap_df.iloc[:top_n_institutions, :top_n_targets]

sns.heatmap(
    joint_heatmap_df, ax=grid.ax_joint, cbar=False, xticklabels=True, yticklabels=True
)

sns.barplot(joint_heatmap_df.sum(axis=0), ax=grid.ax_marg_x, width=1)

sns.barplot(joint_heatmap_df.sum(axis=1), ax=grid.ax_marg_y, width=1, orient="h")

# Offset value (just half an unit)
_off = 0.5

# Fix x
_xmin, _xmax = grid.ax_joint.get_xlim()
grid.ax_joint.set_xlim(_xmin + _off, _xmax + _off)
for bar in grid.ax_marg_x.containers[0]:
    bar.set_x(bar.get_x() + _off)

# Fix y
_ymin, _ymax = grid.ax_joint.get_ylim()
grid.ax_joint.set_ylim(_ymin + _off, _ymax + _off)
for bar in grid.ax_marg_y.containers[0]:
    bar.set_y(bar.get_y() + _off)

grid.ax_joint.set_xlabel(
    f"Top {top_n_targets} Most Frequently Studied Targets (ordered)"
)
grid.ax_joint.set_ylabel(
    f"Top {top_n_institutions} Most Frequent Institutions (ordered)"
)

plt.show()
```
